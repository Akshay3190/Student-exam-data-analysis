```python
plt.figure (figsize = (3.5,3.5))
ax = sns.countplot (data = df, x = 'Gender')
plt.title ('Gender Distribution')
ax.bar_label(ax.containers[0])
plt.show ()
```

```python
#With  the help of above graph we have analysed that the number of female is larger than male in the data.
```

```python
gb = df.groupby ('ParentEduc').agg({'MathScore' : 'mean', 'ReadingScore' : 'mean', 'WritingScore' : 'mean'})
print (gb)
```

```python
plt.figure (figsize = (4,4))
sns.heatmap (gb)
plt.show ()
```

```python
plt.figure (figsize = (4,4))
sns.heatmap (gb, annot = True)
plt.title ('Parent education Vs Children Score')
plt.show ()
```

```python
#From the above graph we can conclude that parent's education have great impact on their chindren's scores.
```

```python
gb1 = df.groupby ('ParentMaritalStatus').agg({'MathScore' : 'mean', 'ReadingScore' : 'mean', 'WritingScore' : 'mean'})
print (gb1)
```

```python
plt.figure (figsize = (4,4))
sns.heatmap (gb1, annot = True)
plt.title ('Parent Marital Status Vs Children Score')
plt.show ()
```

```python
#From the above analysis we can say that parent's marital status have no major/negligible impact on their children's score.
```
